```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 파일 읽기
file_path = './mm_scale.csv'
data = pd.read_csv(file_path)

# 단지코드 저장 후, 특성 데이터만 추출
codes = data['단지코드']
features = data.drop(columns=['단지코드'])

# 코사인 유사도 계산
similarity_matrix = cosine_similarity(features)

# 코사인 유사도 결과를 데이터프레임으로 변환
similarity_df = pd.DataFrame(similarity_matrix, index=codes, columns=codes)

# 특정 단지와 가장 유사한 단지 찾기 (예: 첫 번째 단지)
target_code = codes.iloc[0]  # 첫 번째 단지코드
similarities_to_target = similarity_df[target_code]

# 자기 자신 제외 (유사도 1인 자기 자신 제거)
similarities_to_target = similarities_to_target.drop(target_code)

# 가장 유사한 단지 찾기
most_similar_code = similarities_to_target.idxmax()
most_similar_value = similarities_to_target.max()

# 결과 출력
print(f"'{target_code}' 단지와 가장 유사한 단지: {most_similar_code}")
print(f"유사도: {most_similar_value}")

# 유사도 행렬 저장 (선택 사항)
# similarity_df.to_csv('./cosine_similarity_matrix.csv')
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# 파일 읽기
file_path = './mm_scale.csv'
data = pd.read_csv(file_path)

# 단지코드 저장 후, 특성 데이터만 추출
codes = data['단지코드']
features = data.drop(columns=['단지코드'])

# 코사인 유사도 계산
similarity_matrix = cosine_similarity(features)

# 코사인 유사도 결과를 데이터프레임으로 변환
similarity_df = pd.DataFrame(similarity_matrix, index=codes, columns=codes)

# 특정 단지와 가장 유사한 단지 찾기 (예: 첫 번째 단지)
target_code = codes.iloc[0]  # 첫 번째 단지코드
similarities_to_target = similarity_df[target_code]

# 자기 자신 제외 (유사도 1인 자기 자신 제거)
similarities_to_target = similarities_to_target.drop(target_code)

# 상위 5개 유사한 단지 찾기
most_similar_codes = similarities_to_target.nlargest(5)

# 결과 출력
print(f"'{target_code}' 단지와 가장 유사한 단지들:")
print(most_similar_codes)

# 유사도 행렬 저장 (선택 사항)
# similarity_df.to_csv('./cosine_similarity_matrix.csv')
```
