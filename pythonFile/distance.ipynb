```python
import pandas as pd

# CSV 파일에서 데이터 읽기
data = pd.read_csv("C:/Users/user/Downloads/단지코드_아파트명_요소유형_요소명_거리(m).csv")

# 결과 저장용 리스트
results = []

# 데이터 처리
# 데이터 처리
for _, row in data.iterrows():
    apartment_id = row["아파트명"]
    
    # 병원의 거리 확인
    if row["요소 유형"] == "기관명":
        # 거리 데이터를 정수로 변환
        try:
            hospital_distance = int(row["거리 (m)"])
        except (ValueError, AttributeError):
            print(f"잘못된 거리 값: {row['거리 (m)']} (아파트 ID: {apartment_id})")
            continue

        # 5km(5000m) 이내의 병원만 처리
        if hospital_distance <= 5000:
            # 결과 저장 (5km 이내 병원만 저장)
            results.append({
                "아파트명": apartment_id,
                "병원명": row["요소명"],
                "거리 (m)": hospital_distance
            })

# 결과를 데이터프레임으로 변환 후 CSV 파일로 저장
results_df = pd.DataFrame(results)
results_df.to_csv("hospitals_within_5km.csv", index=False)

print("병원 정보가 'hospitals_within_5km.csv'에 저장되었습니다.")
```
