```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv('./마이홈_merged_with_geo.csv', encoding='utf-8-sig')  # 파일 경로와 인코딩 확인
df.head()
```

```python
# 구 이름 리스트
gu_list = [
    '강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구',
    '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구',
    '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구'
]

# '단지명'에 '서울특별시'와 구 이름이 포함된 경우 수정
df['단지명'] = df.apply(
    lambda row: row['주소'].replace('서울특별시', '').strip() 
    if any(f"서울특별시 {gu}" in row['단지명'] for gu in gu_list) else row['단지명'], 
    axis=1
)

# 결과 확인
print(df[['단지명', '주소']].head())

df.to_csv('마이홈_mergefinal_단지명 변환.csv', index=False, encoding='utf-8-sig')
```
